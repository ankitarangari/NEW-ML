### Finding Outliers and Treating them

```python
var1=['Occupation','Product_Category_1','Product_Category_2']
i=1
for j in var1:
    plt.figure(i)
    sns.boxplot(data[j])
    i=i+1
```

```python
def replace_outliers(car1):
    IQR=data[car1].quantile(0.75)-data[car1].quantile(0.25)
    lower_val=data[car1].quantile(0.25)-(IQR*1.5)
    upper_val=data[car1].quantile(0.75)+(IQR*1.5)
    data.loc[data[car1]>=upper_val,car1]=upper_val
    data.loc[data[car1]<=lower_val,car1]=lower_val
```

```python
for k in var1:
    replace_outliers(k)
```

```python
f=1
for m in var1:
    plt.figure(f)
    sns.boxplot(data[m])
    f+=1
```
